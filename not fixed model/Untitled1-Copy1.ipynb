```python
def calculate_vis_hitting_point(arr):
    __hitting_point = -45
    if (-50 <= float(np.min(arr)) <= -45):
        index = np.argmin(arr)
        row = int((index-0)/3)
        w1 = (arr[row,1]-__hitting_point) / (arr[row,1]-arr[row-1,1])
        __vis_hitting_point = (1-w1)*arr[row,:] + w1*arr[row-1,:]
    elif (-45 < float(np.min(arr)) < -40):
        index = np.argmin(arr)
        row = int((index-0)/3)
        w1 = (arr[row,1]-__hitting_point) / (arr[row-1,1]-arr[row,1])
        __vis_hitting_point = arr[row,:] - w1*(arr[row-1,:]-arr[row,:])
    print("vis hitting point = ", __vis_hitting_point)
```

```python
calculate_vis_hitting_point(tmp.reshape(int(tmp.shape[0]/3),3))
```

```python
def set_data(arr):
    if np.min(arr) <= -45:
        t = arr[np.argmin(arr)-4:np.argmin(arr)-1,]
        print(t)
    else:
        t = arr[np.argmin(arr)-4:np.argmin(arr)+2,]
        print(t)
    tt = sequence.pad_sequences(t.reshape(1,2*3), maxlen=27, padding='post', dtype='float32')
    print("ttt = \n ", tt)
    return tt
```

```python
tt = set_data(tmp)
```

```python
pred = model.predict(tt.reshape(1,9,3))
pred
```
