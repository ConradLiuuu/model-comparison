```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
err = pd.read_csv('./data/all/model/back5_model_error_distribution.csv', header=None)
err = err.fillna(0)
err = np.array(err)
err.shape
```

```python
top5_err = pd.read_csv('./data/all/model/top5_model_error_distribution.csv', header=None)
left5_err = pd.read_csv('./data/all/model/left5_model_error_distribution.csv', header=None)
right5_err = pd.read_csv('./data/all/model/right5_model_error_distribution.csv', header=None)
back5_err = pd.read_csv('./data/all/model/back5_model_error_distribution.csv', header=None)
```

```python
top5_err = np.array(top5_err)
left5_err = np.array(left5_err)
right5_err = np.array(right5_err)
back5_err = np.array(back5_err)
print(top5_err.shape, left5_err.shape, right5_err.shape, back5_err.shape)
```

```python
vel5 = [top5_err, left5_err, right5_err, back5_err]
```

```python
plt.figure(figsize=(9,6))
boxplot_dict = plt.boxplot(vel5)
for i in boxplot_dict['medians']:
    x, y = i.get_xydata()[0]
    plt.text(x, y, '%.3f' % y,horizontalalignment='right',verticalalignment='center')
    #print(x)
for i in boxplot_dict['boxes']:
    x, y = i.get_xydata()[1]
    plt.text(x, y, '%.3f' % y,horizontalalignment='left',verticalalignment='top')
    x2, y2 = i.get_xydata()[2]
    plt.text(x2, y2, '%.3f' % y2,horizontalalignment='left',verticalalignment='bottom')
for i in boxplot_dict['caps']:
    x, y = i.get_xydata()[0]
    plt.text(x, y, '%.3f' % y,horizontalalignment='right',verticalalignment='center')
```
