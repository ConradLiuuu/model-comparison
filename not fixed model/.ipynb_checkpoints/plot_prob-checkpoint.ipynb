```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
name = 'back6'
```

```python
ideal = np.array(pd.read_csv('/home/lab606a/Documents/4Journal_20200629/result/not fixed/'+name+'/ideal/prob.csv'))
cla = np.array(pd.read_csv('/home/lab606a/Documents/4Journal_20200629/result/not fixed/'+name+'/with classification/prob.csv'))
wocla = np.array(pd.read_csv('/home/lab606a/Documents/4Journal_20200629/result/not fixed/'+name+'/without classification/prob.csv'))
```

```python
x = np.arange(0,21,1)
x
```

```python
plt.figure(figsize=(12,9))
plt.plot(x, wocla, marker='o', color='r', label='model A')
plt.plot(x, cla, marker='D', color='b', label='model B')
plt.plot(x, ideal, marker='s', color='g', label='model C')

plt.legend()
plt.grid(True)
plt.xlabel('update times', fontsize=24, fontname='FreeSerif')
plt.ylabel('probability (%)', fontsize=24, fontname='FreeSerif')
plt.xticks(x, fontsize=20, fontname='FreeSerif')
plt.yticks(fontsize=20, fontname='FreeSerif')
plt.tight_layout()
plt.savefig('/home/lab606a/Documents/4Journal_20200629/result/not fixed/'+name+'/9ball_'+name+'.png')
```
