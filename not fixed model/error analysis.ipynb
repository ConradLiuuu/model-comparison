```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
without_classification = pd.read_csv('/home/lab606a/Documents/comp/without classification/without_classification.csv')
```

```python
without_classification = np.array(without_classification)
without_classification.shape
```

```python
x = np.arange(0,without_classification.shape[1])
x
```

```python
#plt.subplot(3,1,1)
plt.errorbar(x, without_classification[1,:].reshape(21,), without_classification[4,:].reshape(21,), linestyle='None', marker='o', color='g', label='t error')
```

```python
ideal = pd.read_csv('/home/lab606a/Documents/comp/ideal/ideal.csv')
```

```python
ideal = np.array(ideal)
ideal.shape
```

```python
plt.errorbar(x, ideal[1,:].reshape(21,), ideal[4,:].reshape(21,), linestyle='None', marker='o', color='g', label='t error')
```

```python
with_classification = pd.read_csv('/home/lab606a/Documents/comp/with classification/with_classification.csv')
```

```python
with_classification = np.array(with_classification)
with_classification.shape
```

```python
plt.errorbar(x, with_classification[1,:].reshape(21,), with_classification[4,:].reshape(21,), linestyle='None', marker='o', color='g', label='t error')
```

```python
#plt.subplot(3,1,1)
plt.figure(figsize=(12,9))
plt.errorbar(x, without_classification[0,:].reshape(21,), without_classification[3,:].reshape(21,), linestyle='None', marker='o', color='r', alpha=1, label='model A')
plt.errorbar(x, with_classification[0,:].reshape(21,), with_classification[3,:].reshape(21,), linestyle='None', marker='D', color='b', alpha=1, label='model B')
plt.errorbar(x, ideal[0,:].reshape(21,), ideal[3,:].reshape(21,), linestyle='None', marker='s', color='g', alpha=1, label='model C')
plt.grid(True)
plt.legend()
plt.tight_layout(pad=5)
plt.title('t error bar', fontsize=20)
plt.xlabel('Update times', fontsize=20)
plt.ylabel('error (sec)', fontsize=20)
plt.xticks(np.arange(0,21), fontsize=16)
plt.yticks(fontsize=16)
plt.savefig('/home/lab606a/Documents/comp/t error bar.png')
plt.savefig('/home/lab606a/Documents/comp/t error bar.svg')
```
