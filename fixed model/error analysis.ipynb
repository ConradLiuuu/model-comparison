```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
type_ = 'back6'
```

```python
without_classification = pd.read_csv('/home/lab606a/Documents/4Journal_20200629/result/fixed/'+type_+'/without classification/without_classification.csv')
without_classification = np.array(without_classification)
```

```python
ideal = pd.read_csv('/home/lab606a/Documents/4Journal_20200629/result/fixed/'+type_+'/ideal/ideal.csv')
ideal = np.array(ideal)
```

```python
with_classification = pd.read_csv('/home/lab606a/Documents/4Journal_20200629/result/fixed/'+type_+'/with classification/with_classification.csv')
with_classification = np.array(with_classification)
```

```python
x = np.arange(0,without_classification.shape[1])
x
```

```python
#plt.subplot(3,1,1)
plt.figure(figsize=(12,9))
plt.errorbar(x, without_classification[0,:].reshape(21,), without_classification[3,:].reshape(21,), linestyle='None', marker='o', color='r', alpha=1, label='model A')
plt.errorbar(x, with_classification[0,:].reshape(21,), with_classification[3,:].reshape(21,), linestyle='None', marker='D', color='b', alpha=1, label='model B')
plt.errorbar(x, ideal[0,:].reshape(21,), ideal[3,:].reshape(21,), linestyle='None', marker='s', color='g', alpha=1, label='model C')
plt.grid(True)
plt.legend()
plt.tight_layout(pad=5)
plt.title('t error bar', fontsize=20, fontname='FreeSerif')
plt.xlabel('update times', fontsize=20, fontname='FreeSerif')
plt.ylabel('error (ms)', fontsize=20, fontname='FreeSerif')
plt.xticks(np.arange(0,21), fontsize=16, fontname='FreeSerif')
plt.yticks(np.array([-33.34, -16.67, 0, 16.67, 33.34]), fontsize=16)
plt.yticks(fontsize=16, fontname='FreeSerif')

plt.savefig('/home/lab606a/Documents/4Journal_20200629/result/fixed/'+type_+'/t error bar.png')
plt.savefig('/home/lab606a/Documents/4Journal_20200629/result/fixed/'+type_+'/t error bar.svg')
```
